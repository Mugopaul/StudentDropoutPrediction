```python
# Pie chart to display the amount (percentage) distribution of the Target Variable

labels = 'Graduate', 'Dropout', 'Enrolled'
sizes = [data.Target[data['Target']=='Graduate'].count(), data.Target[data['Target']=='Dropout'].count(), data.Target[data['Target']=='Enrolled'].count()]
explode = (0, 0.1, 0.1)
fig1, ax1 = plt.subplots(figsize=(12, 8))
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')
plt.title("Proportion of Target variable", size = 10)
plt.show()
```

## Evaluating the Discrete Random Variables/ Categorical Data

```python
cols_ = ["Mother's occupation",
 "Father's occupation",
 'Marital status',
 'Application order',
 'Age at enrollment',
 "Father's qualification",
 "Mother's qualification",
 'Tuition fees up to date',
 'Gender',
  "Nacionality",
 'Displaced',
 'Debtor',
 "Scholarship holder",
 'Application mode',
 'Course']


# Define the number of subplots
num_plots = len(cols_)

# Create figure and axes
fig, ax = plt.subplots(nrows=num_plots, ncols=1, figsize=(20, 5 * num_plots))  

# Flatten ax in case it's not a 1D array
ax = ax.flatten()

# Loop through columns and plot
for ind, col in enumerate(cols_):
    sns.countplot(x=col, data=data, ax=ax[ind])
    ax[ind].set_title(f"Distribution of {col}", fontsize=14)
    ax[ind].tick_params(axis='x', rotation=45)  # Rotate x-axis labels for readability

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Let's Analyze the impact of each of the above variable on the Target Variable

import math

# Variables to analyze
cols = ['Application order', 'Tuition fees up to date', 'Gender', 
        'Displaced', 'Debtor', 'Scholarship holder']

# Define grid size dynamically
num_vars = len(cols)
num_cols = 3  # Number of columns in the subplot grid
num_rows = math.ceil(num_vars / num_cols)  # Calculate required rows

# Create figure and axes
fig, ax = plt.subplots(num_rows, num_cols, figsize=(20, 5 * num_rows))

# Flatten ax if needed (for easier indexing)
ax = ax.flatten()

# Loop through variables and plot
for i, col in enumerate(cols):
    sns.countplot(x=col, hue='Target', data=data, palette='Set2', ax=ax[i])
    ax[i].set_title(f"Impact of {col} on Target", fontsize=14)
    ax[i].tick_params(axis='x', rotation=45)  # Rotate labels for readability

# Remove unused subplots (if any)
for j in range(i + 1, len(ax)):
    fig.delaxes(ax[j])  # Remove empty subplot spaces

# Adjust layout
plt.tight_layout()
plt.show()
```
